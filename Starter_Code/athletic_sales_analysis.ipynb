```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_data_rows_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis='rows', join='inner')
joined_data_rows_df.reset_index(drop=True)
```

```python
# Check if any values are null.
null_values = joined_data_rows_df.isna().sum()
print(null_values)
```

```python
# Check the data type of each column
joined_data_rows_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_data_rows_df['invoice_date'] = pd.to_datetime(joined_data_rows_df['invoice_date'], format='%Y-%m-%d', utc=True)
```

```python
# Confirm that the "invoice_date" data type has been changed.
joined_data_rows_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
new_units_sold_df = joined_data_rows_df[['region','state','city','units_sold']]
grouped_units_total_df = new_units_sold_df.groupby(['region','state','city']).agg({'units_sold':'sum'})
renamed_grouped_units_total_df = grouped_units_total_df.rename(columns={'units_sold':'Total_Products_Sold'})

# Show the top 5 results.
renamed_grouped_units_total_df.sort_values(by='Total_Products_Sold', ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table_units_total_df = pd.pivot_table(joined_data_rows_df, 
                                            index=['region','state','city'], 
                                            values='units_sold',
                                            aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
renamed_pivot_units_total_df = pivot_table_units_total_df.rename(columns={'units_sold':'Total_Products_Sold'})

# Show the top 5 results.
renamed_pivot_units_total_df.sort_values(by='Total_Products_Sold', ascending=False).head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
new_sales_total_df = joined_data_rows_df[['region','state','city','total_sales']]
grouped_sales_total_df = new_sales_total_df.groupby(['region','state','city']).agg({'total_sales':'sum'})
renamed_grouped_sales_total_df = grouped_sales_total_df.rename(columns={'total_sales':'Total Sales'})

# Show the top 5 results.
renamed_grouped_sales_total_df.sort_values(by='Total Sales', ascending=False).head()

```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_table_sales_total_df = pd.pivot_table(joined_data_rows_df, 
                                            index=['region','state','city'], 
                                            values='total_sales',
                                            aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
renamed_pivot_sales_total_df = pivot_table_sales_total_df.rename(columns={'total_sales':'Total Sales'})

# Show the top 5 results.
renamed_pivot_sales_total_df.sort_values(by='Total Sales', ascending=False).head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
joined_data_rows_df.value_counts('product')
womens_athletic_footwear_df = joined_data_rows_df.loc[(joined_data_rows_df['product'] == "Women's Athletic Footwear")]
womens_athletic_footwear_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
units_sold_womens_athletic_footwear_df = womens_athletic_footwear_df[['retailer','region','state','city','units_sold']]
grouped_units_sold_womens_athletic_footwear_df = units_sold_womens_athletic_footwear_df.groupby(['retailer','region','state','city']).agg({'units_sold':'sum'})
renamed_grouped_units_sold_womens_athletic_footwear_df = grouped_units_sold_womens_athletic_footwear_df.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
renamed_grouped_units_sold_womens_athletic_footwear_df.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
units_sold_pivot_table_womens_athletic_footwear_df = pd.pivot_table(womens_athletic_footwear_df, 
                                            index=['retailer','region','state','city'], 
                                            values='units_sold',
                                            aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
renamed_units_sold_pivot_table_womens_athletic_footwear_df = units_sold_pivot_table_womens_athletic_footwear_df.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
renamed_units_sold_pivot_table_womens_athletic_footwear_df.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
invoice_total_pivot_df = pd.pivot_table(womens_athletic_footwear_df, 
                                            index='invoice_date', 
                                            values='total_sales',
                                            aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
rename_invoice_total_pivot_df = invoice_total_pivot_df.rename(columns={'total_sales': 'Total Sales'})

# Show the table.
rename_invoice_total_pivot_df
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_invoice_total_df = rename_invoice_total_pivot_df.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
daily_invoice_total_df.sort_values(by='Total Sales', ascending=False).head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_invoice_total_df = rename_invoice_total_pivot_df.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
weekly_invoice_total_df.sort_values(by='Total Sales', ascending=False).head(10)
```
